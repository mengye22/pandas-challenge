```python
# Pandas Homework City School
# By Mengye Wei
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Rename the column names
school_data = school_data.rename(columns={'school_name' : 'School_Name',
                                          'type' : 'Type',
                                          'size' : 'Size',
                                          'budget' : 'Budget'})
school_data.head()
```

```python
# Rename the column names
student_data = student_data.rename(columns={'student_name' : 'Student_Name',
                                            'gender' : 'Gender',
                                            'grade' : 'Grade',
                                            'school_name' : 'School_Name',
                                            'reading_score' : 'Reading_Score', 
                                            'math_score' : 'Math_Score'})
student_data.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["School_Name", "School_Name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total or average 
total_schools = school_data["School_Name"].count()
total_students = student_data["Student_Name"].count()
total_budget = school_data["Budget"].sum()
avg_math_score = student_data["Math_Score"].mean()
avg_reading_score = student_data["Reading_Score"].mean()

# Frist locate the students score 70/70+ in the dataset
# Then divide the count by the total number to find the percentage
pass_math = student_data.loc[student_data["Math_Score"] > 69]
pass_math_percent = pass_math["Math_Score"].count()/total_students

pass_reading = student_data.loc[student_data["Reading_Score"] > 69]
pass_reading_percent = pass_reading["Reading_Score"].count()/total_students

# Find the overall pass rate by average the average of math and reading 
# For rhe overall_pass_percent here, I use the pass percentage in both math and
# reading, but not average the total score, hence, and a bit difference
# on my datasets
overall_pass_percent = (pass_math_percent+pass_reading_percent)/2

# Put all the datasets into data frame with currency and percentage format

district_summary = pd.DataFrame({"Total Schools":[total_schools],
                           "Total Students":["{:,}".format(total_students)],
                           "Total Budget":["${:,.2f}".format(total_budget)],
                           "Average Math Score":[avg_math_score],
                           "Average Reading Score":[avg_reading_score],
                           "Passing Math":["{:.3%}".format(pass_math_percent)],
                           "Passing Reading":["{:.3%}".format(pass_reading_percent)],
                           "Overall Passing Rate":["{:.3%}".format(overall_pass_percent)]})
# Dispaly the Data
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Display the first five roads of 
school_data_complete.head()
```

```python
# Set school name as index by school type
school_type = school_data.set_index(["School_Name"])["Type"]

# Calculate the total students in each school
total_student = school_data_complete["School_Name"].value_counts()

# Find the budget per school and per student
total_school_budget = school_data_complete.groupby(["School_Name"]).mean()["Budget"]
per_student_budget = total_school_budget / total_student
# Find the average score by schools
school_avg_math =  school_data_complete.groupby(["School_Name"]).mean()["Math_Score"]
school_avg_reading =  school_data_complete.groupby(["School_Name"]).mean()["Reading_Score"]
# Find the percentage of students who score 70/70+ 
student_passing_math = school_data_complete.loc[school_data_complete["Math_Score"] > 69].groupby(["School_Name"]).count()["Student_Name"]
student_passmath_rate = student_passing_math / total_student
student_passing_reading = school_data_complete.loc[school_data_complete["Reading_Score"] > 69].groupby(["School_Name"]).count()["Student_Name"]
student_passreading_rate = student_passing_reading / total_student
# Average score by school
school_overall_passing = (student_passmath_rate + student_passreading_rate) / 2

# Dataframe the datasets calculated above
school_summary = pd.DataFrame({"School Type":school_type,
                              "Total Students":total_student,
                              "Total School Budget":total_school_budget.map("${:,.2f}".format),
                              "Per Student Budget":per_student_budget.map("${:,.2f}".format),
                              "Average Math Score":school_avg_math,
                              "Average Reading Score":school_avg_reading,
                              "Passing Math":student_passmath_rate.map("{:.3%}".format),
                              "Passing Reading":student_passreading_rate.map("{:.3%}".format),
                              "Overall Passing Rate":school_overall_passing.map("{:.3%}".format)})

# Display the data
school_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# Display the top performing schools by sortting descending the overall passing rate
top_perform = school_summary.sort_values("Overall Passing Rate",ascending = False)
top_perform.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
# Display the bottom performing schools by sorrting ascending the overall passing rate
bottom_perform = school_summary.sort_values("Overall Passing Rate",ascending = True)
bottom_perform.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# First, find the grade level in school_data_complete
# Then, groupby school name
# Last, take the average of the math score
ninth_grade_math = school_data_complete[school_data_complete["Grade"] == "9th"].groupby("School_Name").mean()["Math_Score"]
tenth_grade_math = school_data_complete[school_data_complete["Grade"] == "10th"].groupby("School_Name").mean()["Math_Score"]
eleventh_grade_math = school_data_complete[school_data_complete["Grade"] == "11th"].groupby("School_Name").mean()["Math_Score"]
twelfth_grade_math = school_data_complete[school_data_complete["Grade"] == "12th"].groupby("School_Name").mean()["Math_Score"]

# Put datas in the datafram and display it
math_score_by_grade = pd.DataFrame({"9th":ninth_grade_math,
                                   "10th":tenth_grade_math,
                                   "11th":eleventh_grade_math,
                                   "12th":twelfth_grade_math})
math_score_by_grade
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
# First, find the grade level in school_data_complete
# Then, groupby school name
# Last, take the average of the math score
ninth_grade_reading = school_data_complete[school_data_complete["Grade"] == "9th"].groupby("School_Name").mean()["Reading_Score"]
tenth_grade_reading = school_data_complete[school_data_complete["Grade"] == "10th"].groupby("School_Name").mean()["Reading_Score"]
eleventh_grade_reading = school_data_complete[school_data_complete["Grade"] == "11th"].groupby("School_Name").mean()["Reading_Score"]
twelfth_grade_reading = school_data_complete[school_data_complete["Grade"] == "12th"].groupby("School_Name").mean()["Reading_Score"]

# Put datas in the datafram and display it
reading_score_by_grade = pd.DataFrame({"9th":ninth_grade_reading,
                                   "10th":tenth_grade_reading,
                                   "11th":eleventh_grade_reading,
                                   "12th":twelfth_grade_reading})
reading_score_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create bins and bins name to put datas in
spending_bins = [0, 584, 614, 644, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_summary["Spending Ranges(Per Student)"] = pd.cut(per_student_budget, spending_bins,
                                                        labels=group_names)

# Remove the percentage sign in the data for further calculation
school_summary["Passing Math"] = school_summary["Passing Math"].replace('%', '', regex=True).astype('float')
school_summary["Passing Reading"] = school_summary["Passing Reading"].replace('%', '', regex=True).astype('float')
school_summary["Overall Passing Rate"] = school_summary["Overall Passing Rate"].replace('%', '', regex=True).astype('float')

# Create a new variable for groupby the Spending Range and display the dataset
scores_bySchool_spending = school_summary.groupby("Spending Ranges(Per Student)")
scores_bySchool_spending["Average Math Score","Average Reading Score",
                         "Passing Math","Passing Reading", "Overall Passing Rate"].mean()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# create bins and bins name for school size
size_bins = [0, 1000, 2000, 5000]
group_size = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_summary["School Size"] = pd.cut(total_student, size_bins,
                                                        labels=group_size)

# Create a new variable for groupby the school size and display the dataset
scores_bySchool_size = school_summary.groupby("School Size")
scores_bySchool_size["Average Math Score","Average Reading Score",
                         "Passing Math","Passing Reading", "Overall Passing Rate"].mean()
```

## Scores by School Type

* Perform the same operations as above, based on school type.

```python
# Since we only have two school types, we can groupby here. No need new bins
scores_bySchool_type = school_summary.groupby("School Type")
scores_bySchool_type["Average Math Score","Average Reading Score",
                         "Passing Math","Passing Reading", "Overall Passing Rate"].mean()
```
